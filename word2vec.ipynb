```python
import random
from utils.treebank import StanfordSentiment

import numpy as np
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt

import torch
import torch.nn
import torch.nn.functional as F


# Reset the random seed to make sure that everyone gets the same results
random.seed(314)
dataset = StanfordSentiment()
tokens = dataset.tokens()
nWords = len(tokens)

# We are going to train 10-dimensional vectors for this assignment
dimVectors = 10
# Context size
C = 5

device = "cuda" if torch.cuda.is_available() else "cpu"

g = torch.Generator().manual_seed(2147483647)
```

```python
#c = torch.nn.Embedding(nWords, dimVectors, device=device)
#o = torch.nn.Embedding(nWords, dimVectors, device=device)
U = torch.randn(nWords, dimVectors, device=device, generator=g, requires_grad=True)
V = torch.randn(nWords, dimVectors, device=device, generator=g, requires_grad=True)
```

```python
c = torch.tensor(1)
o = torch.tensor(5)

cenc = F.one_hot(c, num_classes=nWords).float()
oenc = F.one_hot(o, num_classes=nWords).float()
#logits = U[c] @ V[o]
#counts = logits.exp()
#total = (U.t() @ V).exp().sum(dim=1)
#print(total[c])
vc = cenc @ V
uwc = U @ vc.t()
u_sum = uwc.exp().sum()

uo = oenc @ U
logits = uo @ vc

prob = logits.exp() / u_sum

print(prob)

loss = -torch.log(prob)
print(loss)
```

```python
window_size = 5
for i in range(1000):
    centerWord, context = dataset.getRandomContext(window_size)

    if centerWord not in tokens:
        continue

    for outer in context:
        if outer not in tokens:
            continue
        c = torch.tensor(tokens[centerWord])
        o = torch.tensor(tokens[outer])
        cenc = F.one_hot(c, num_classes=nWords).float()
        oenc = F.one_hot(o, num_classes=nWords).float()
        #logits = U[c] @ V[o]
        #counts = logits.exp()
        #total = (U.t() @ V).exp().sum(dim=1)
        #print(total[c])
        vc = cenc @ V
        uwc = U @ vc.t()
        u_sum = uwc.exp().sum()

        uo = oenc @ U
        logits = uo @ vc

        prob = logits.exp() / u_sum

        loss = -torch.log(prob)
        U.grad = None
        V.grad = None
        loss.backward()
        V.data -= 0.01 * V.grad.data
        U.data -= 0.01 * U.grad.data
        print(loss)
```
